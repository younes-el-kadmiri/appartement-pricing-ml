```python
import pandas as pd
import numpy as np

def clean_data(df):
    # Nettoyer la colonne 'price' : enlever tout ce qui n'est pas chiffre ou point
    df['price'] = df['price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
    # Supposons que tu as filtr√© le DataFrame :
    df = df[df['price'] != '']
    # Assure-toi une copie compl√®te avant la modification :
    df = df.copy()
    # Puis convertis la colonne
    df['price'] = df['price'].astype(float)


    
    # Supprimer les lignes o√π price <= 0 (valeurs invalides)
    df = df[df['price'] > 0]

    # Supprimer les lignes o√π surface_area est manquante ou <= 0
    df = df[df['surface_area'].notna()]
    df = df[df['surface_area'] > 0]

    # Harmoniser noms de villes arabe ‚Üí fran√ßais
    arabic_to_french = {'ÿßŸÑÿØÿßÿ± ÿßŸÑÿ®Ÿäÿ∂ÿßÿ°': 'Casablanca', 'ÿßŸÑÿ±ÿ®ÿßÿ∑': 'Rabat'}
    if 'city_name' in df.columns:
        df['city_name'] = df['city_name'].replace(arabic_to_french)
        df['city_name'] = df['city_name'].fillna('Unknown')

    # Supprimer outliers avec m√©thode IQR sur price et surface_area
    for col in ['price', 'surface_area']:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    # Remplir valeurs manquantes dans d'autres colonnes num√©riques avec m√©diane
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
    for col in numeric_cols:
        if df[col].isnull().any():
            median = df[col].median()
            df.loc[:, col] = df.loc[:, col].fillna(median)



    # Remplir valeurs manquantes dans colonnes cat√©goriques avec 'Unknown'
    cat_cols = df.select_dtypes(include=['object']).columns.tolist()
    for col in cat_cols:
        df.loc[:, col] = df.loc[:, col].fillna('Unknown')

    df.reset_index(drop=True, inplace=True)
    return df


if __name__ == "__main__":
    # Charger les donn√©es
    df = pd.read_csv('/mnt/c/Windows/system32/appartement-pricing-ml/data/appartements_data_db.csv')
    print(f"Avant nettoyage : {df.shape}")

    # Nettoyer
    df_clean = clean_data(df)
    print(f"Apr√®s nettoyage : {df_clean.shape}")

    # Afficher r√©sum√©
    print(df_clean.info())
    print(df_clean.describe())

    # Exemple de visualisation (corr√©lation)
    import matplotlib.pyplot as plt
    import seaborn as sns

    sns.set(style='whitegrid')
    plt.figure(figsize=(10, 6))
    sns.heatmap(df_clean.corr(numeric_only=True), annot=True, cmap='coolwarm')
    plt.title("Matrice de Corr√©lation apr√®s nettoyage")
    plt.show()
```
